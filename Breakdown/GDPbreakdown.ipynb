```python
import requests
from pprint import pprint
import pandas as pd
from pathlib import Path
```

```python
countries_file = Path("../10countries.csv")
```

```python
countries_file_df = pd.read_csv(countries_file)
```

```python
countries_file_df
```

```python
countries_file_df.dtypes
```

```python
cleaned_df = countries_file_df[["series_id", "country_code","year","value"]]
cleaned_df
```

```python
cleaned_df = cleaned_df[cleaned_df["year"] >= 2013]
cleaned_df



```

```python
all_years = sorted(cleaned_df['year'].unique())
all_years
```

```python
gdp_wide = cleaned_df.pivot(index='year', columns='country_code', values='value').reindex(all_years)
gdp_wide
```

```python
cleaned_df.to_csv("../Resources/cleaned_df.csv", index=False, header=True)
gdp_wide.to_excel("../Resources/gdp_data.xlsx")
```
